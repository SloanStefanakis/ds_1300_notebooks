```python
ax = mvps.plot.scatter(x='PER', y="WS", color='Red',label="WS", marker='^')
plt = mvps.plot.scatter(x='PER', y='VORP', label="VORP", ax=ax, xlabel="PER", ylabel="VORP & WS", title='PER WITH V.O.R.P. to Win Shares',figsize=(15,5))
plt.figure.savefig('assignment.pdf')
```

```python
ax = mvps.plot.scatter(x='VORP', y="WS", color='Red',label="WS", marker='^')
mvps.plot.scatter(x='VORP', y='PER', label="PER", ax=ax, xlabel="VORP", ylabel="PER & WS", title='V.O.R.P. to Win Shares & PER',figsize=(15,5))
```

```python
ax = mvps.plot.scatter(x='WS', y="PER", color='Red',label="PER", marker='^')
mvps.plot.scatter(x='WS', y='VORP', label="WS", ax=ax, xlabel="WS", ylabel="VORP & PER", title='Win Shares to VORP & PER',figsize=(15,5))
```
